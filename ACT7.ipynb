```python
import numpy as np
import pandas as pd 
# import seaborn as sns 
# import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap 
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('dataset\\terrorism.csv', encoding='ISO-8859-1')
df
```

# Crime Type Analysis
<h5>This map highlights severe kidnapping incidents before 2010 that resulted in notable fatalities. Hotspots here likely indicate regions where kidnapping was used not just as a coercive tactic, but as a deadly strategy—possibly in high-conflict zones or areas with weak governance. These may correlate with insurgent or terrorist activities where hostages were executed or harmed during rescue attempts.</h5>

```python
df_kidnap = df[(df['attacktype1_txt'] == 'Hostage Taking (Kidnapping)') &
               (df['iyear'] < 2010) &
               (df['nkill'] >= 3)]

print(f"Filtered dataset size: {len(df_kidnap)} rows")
print(df_kidnap.head(10))

loc_group = df_kidnap.groupby(['latitude', 'longitude']).size().reset_index(name='count')

loc_group['weight'] = (loc_group['count'] - loc_group['count'].min()) / \
                      (loc_group['count'].max() - loc_group['count'].min())

heat_vals = loc_group[['latitude', 'longitude', 'weight']].values.tolist()

map_kidnap = folium.Map(location=[30, 40], zoom_start=4)
HeatMap(heat_vals).add_to(map_kidnap)
map_kidnap.save('kidnapping_heatmap.html')
map_kidnap
```

# Target Type Analysis
<h5>This dataset uncovers large-scale attacks on military targets, suggesting either direct combat actions, base bombings, or highly organized assaults. Concentrated heat in specific regions may point to areas of prolonged military conflict or terrorist campaigns against state forces, especially in zones with civil wars or ideological insurgencies during the late 20th century.</h5>

```python
df_military = df[(df['targtype1_txt'] == 'Military') &
                 (df['iyear'] < 2000) &
                 (df['nkill'] >= 20)]

print(f"Filtered dataset size: {len(df_military)} rows")
print(df_military.head(10))

military_loc = df_military.groupby(['latitude', 'longitude']).size().reset_index(name='count')

military_loc['weight'] = (military_loc['count'] - military_loc['count'].min()) / \
                         (military_loc['count'].max() - military_loc['count'].min())

military_heat = military_loc[['latitude', 'longitude', 'weight']].values.tolist()

map_military = folium.Map(location=[30, 40], zoom_start=4)
HeatMap(military_heat).add_to(map_military)
map_military.save('military_target_heatmap.html')
map_military                         
              
```

# Casualty Analysis
<h5>This general filter highlights globally significant violent events based solely on death toll. It's useful for identifying the most destructive incidents across regions and attack types. The spread of the heatmap may indicate where preventive efforts and global attention were (or should have been) focused. The time constraint also avoids skew from post-2005 trends like ISIS or more recent insurgencies.</h5>

```python
df_cas = df[(df['nkill'] >= 15) & (df['iyear'] < 2005)]

print(f"Filtered dataset size: {len(df_cas)} rows")
print(df_cas.head(10))

cas_group = df_cas.groupby(['latitude', 'longitude']).size().reset_index(name='count')

cas_group['weight'] = (cas_group['count'] - cas_group['count'].min()) / \
                      (cas_group['count'].max() - cas_group['count'].min())
                      
cas_heat = cas_group[['latitude', 'longitude', 'weight']].values.tolist()

map_cas = folium.Map(location=[30, 40], zoom_start=4)
HeatMap(cas_heat).add_to(map_cas)
map_cas.save('medium_casualty_heatmap.html')
map_cas
```
